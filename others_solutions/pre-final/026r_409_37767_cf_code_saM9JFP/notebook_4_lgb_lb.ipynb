```python
from utils import *
from lightgbm import LGBMClassifier
from itertools import product
```

```python
df = aggregate(lb_encode = True)
df = clean_data(df)
```

```python
lgbm_params = {'nthread': 8, 
               'n_estimators': 10000, 
               'learning_rate': 0.02, 
               'num_leaves': 20, 
               'colsample_bytree': 0.5, 
               'subsample': 0.6, 
               'max_depth': 8, 
               'min_split_gain': 0.005, 
               'min_child_weight': 39.3259775, 
              'eval_metric': 'logloss',
              'reg_lambda': 0.03}
logging = setup_logging()
logger = logging.getLogger('LGBM')
```

```python
estimator = LGBMClassifier()
for iteration, random_state in product([1, 2, 3, 4, 5], [11, 111, 786, 101, 5]):
    feature_importance, scor = cv_scores(estimator, df, num_folds = 5, 
                                     params = lgbm_params, stratified = False, random_state = random_state,
                                     train_prediction_file_name = f'../utility/X_train_base_lb_True_{iteration}_{random_state}.csv',
                                     test_prediction_file_name = f'../utility/X_test_base_lb_True_{iteration}_{random_state}.csv',
                                         iteration = f'iter:{iteration}, random states:{random_state}',
                                    logger = logger)
```
