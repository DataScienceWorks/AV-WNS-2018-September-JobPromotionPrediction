```python
from utils import *
from catboost import CatBoostClassifier
from itertools import product
```

```python
df = aggregate(lb_encode = True)
df = clean_data(df)
```

```python
cat_params = {'use_best_model': True,
              'n_estimators': 2500,
            'learning_rate': .02,
            'max_depth': 8,
            'silent': True,
              'loss_function':'Logloss', 
              'eval_metric':'AUC',
}
logging = setup_logging()
logger = logging.getLogger('CatBoost')
```

```python
estimator = CatBoostClassifier()
for iteration, random_state in product([1, 2, 3], [11, 111, 786]):
    feature_importance, scor = cv_scores(estimator, df, num_folds = 5, 
                                     params = cat_params, stratified = False, random_state = random_state,
                                     train_prediction_file_name = f'../utility/X_train_base_cat_lb_false_{iteration}_{random_state}.csv',
                                     test_prediction_file_name = f'../utility/X_test_base_cat_lb_false_{iteration}_{random_state}.csv',
                                         iteration = f'iter:{iteration}, random states:{random_state}',
                                    logger = logger, lgbm_clf = False)
```
